```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
train_data=pd.read_csv("Train.csv")
```

```python
X_train, X_val = train_test_split(train_data, test_size=0.2, random_state=18)
X=X_train.drop("emotions",axis=1)
Y=X_train["emotions"]
X_valid=X_val.drop("emotions",axis=1)
Y_valid=X_val["emotions"]
```

```python
emotions = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'}
Y_plt = Y.replace(emotions)
Y_valid_plt = Y_valid.replace(emotions)

plt.figure(figsize=(10, 4)) 

plt.subplot(1, 2, 1) 
Y_plt.value_counts().plot(kind='bar')
plt.title('Training')

plt.subplot(1, 2, 2)
Y_valid_plt.value_counts().plot(kind='bar',color='orange')
plt.title('Validation')

plt.show()
```
